```python
# General Utilites
import os
import numpy as np
import pandas as pd
import random
import yaml
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [20, 20]
from PIL import Image
from shutil import copyfile
import shutil
from IPython.display import clear_output
```

```python
with open("create_scoring_df.yaml", "r") as input_stream: ### set input files
    settings = yaml.load(input_stream, Loader=yaml.FullLoader)
input_dir = f'{settings["input_dir"]}'
df_output_path = f'{settings["df_output_path"]}'
```

```python
scoring_df = pd.DataFrame({"Filename" : os.listdir(input_dir)})
def get_filepath(row):
    return f"{input_dir}/{row['Filename']}"
scoring_df["Filepath"] = scoring_df.apply(lambda row: get_filepath(row), axis=1)
scoring_df.head(10)
```

```python
status_list = ["Reproductive", "Flowering", "Fruiting", "Budding"]

for status in status_list:
    scoring_df.loc[:,f"{status} Status"] = float("NaN")

print(f"Total Number of imageS: {len(scoring_df)}")


scoring_df.loc[:,"Scored"] = False

```

```python
scoring_df.head(10)
```

```python

scoring_df.to_csv(df_output_path)
```
